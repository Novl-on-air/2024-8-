## 文件参数展示
可以展示config.yaml配置文件

```python
import yaml
```

```python
def duruyaml(config_file_name):
    with open(config_file_name,'r') as f:
        config=yaml.load(f,Loader=yaml.FullLoader)
        return config
config__file=os.path.join(os.getcwd(),"config.yaml")
fconfig=duruyaml(config__file)
print(fconfig)
```

## 文件参数修改
可以修改config.yaml配置文件以改变其参数

```python
def rewrite_yaml(config_file_name，参数名,值):
    with open(config_file_name) as f:
        doc=yaml.safe_loadload(f)
        doc[参数名]=值
    with open(config_file_name，'w') as f:
        yaml.safe_dump(doc,f,default_flow_style=False)
config__file=os.path.join(os.getcwd(),"config.yaml")
参数名=input("参数名\n")
值=input("值")
rewrite_yaml(config__file,参数名,值)
```
